```python
import pandas as pd
import os
import shutil

# Path to the folder containing ticker files
ticker_folder_path = "sec-edgar-filings"

# should create a empty output folder
output_folder = "10-Q"

# Load ticker names from a DataFrame (replace this with your own method)
ticker_df = pd.read_csv("ticker_and_cik.csv")  # Adjust the CSV file path accordingly

```

### renaming the files by extracting from the folder using the tickers and saving all the txt files in one folder

```python
# Loop through each ticker name in the DataFrame
successful_save_count = 0
for ticker_name in ticker_df['ticker']:  # Replace 'ticker_column_name' with your actual column name
    # Construct the path to the 10-Q folder for each ticker
    ticker_10Q_path = os.path.join(ticker_folder_path, ticker_name, "10-Q")

    # Check if the 10-Q folder exists for the current ticker
    if os.path.exists(ticker_10Q_path):
        # Get a list of files in the 10-Q folder
        files_in_10Q = os.listdir(ticker_10Q_path)

        # Loop through each file in the 10-Q folder
        for file_name in files_in_10Q:
            if os.path.isdir(os.path.join(ticker_10Q_path, file_name)):
            # Construct the full path to the subfolder
                subfolder_path = os.path.join(ticker_10Q_path, file_name)

            # Get a list of files within the subfolder
                subfolder_files = os.listdir(subfolder_path)

            # Loop through each file within the subfolder
                for subfile_name in subfolder_files:
                # Check if the file is a .txt file
                    if subfile_name.endswith(".txt"):
                    # Construct the full path to the .txt file
                        txt_file_path = os.path.join(subfolder_path, subfile_name)

                    # Open and read the content of the .txt file
                        with open(txt_file_path, 'r') as txt_file:
                            file_content = txt_file.read()

                        first_line = file_content.split('\n')[0]
                        first_line = first_line.replace('<SEC-DOCUMENT>', ' ').strip()
                    # Construct the new file name (ticker name + first line content)
                        new_file_name = f"{ticker_name}_{first_line}.txt"

                    # Construct the full path to the output folder
                        output_file_path = os.path.join(output_folder, new_file_name)

                    

                    # Write the content to the output file
                        with open(output_file_path, 'w') as output_file:
                            output_file.write(file_content)
                        successful_save_count += 1
        print(f"Files have been copied and renamed successfully for {ticker_name}.")
    else:
        print(f"10-Q folder does not exist for {ticker_name}.")
        
# Display the total count of successful saves
print(f"Total number of tickers successfully saved: {successful_save_count}")
```
